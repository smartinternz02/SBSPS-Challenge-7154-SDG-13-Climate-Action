```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
b=pd.read_excel('Adilabad-1.xlsx')
b
```

```python
plt.figure(figsize=(20,5))
plt.title("Adilabad Rain")
sns.scatterplot(x='Mandal',y='Rain (mm)',data=b)
plt.show()
```

```python
plt.figure(figsize=(27,5))
plt.title("Adilabad Rain")
sns.lineplot(x='Mandal',y='Rain (mm)',data=b)
plt.show()
```

```python
plt.figure(figsize=(26,7))
plt.title("Adilabad Rain")
sns.barplot(x='Mandal',y='Rain (mm)',data=b)
plt.show()
```

```python
plt.figure(figsize=(26,5))
plt.title("Adilabad Min Temp")
sns.barplot(y='Min Temp (°C)',x='Mandal',data=b)
plt.show()
```

```python
plt.figure(figsize=(26,5))
plt.title("Adilabad Min Temp")
sns.lineplot(x='Mandal',y='Min Temp (°C)',data=b)
plt.show()
```

```python
plt.figure(figsize=(26,5))
plt.title("Adilabad Max Temp")
sns.barplot(y='Max Temp (°C)',x='Mandal',data=b)
plt.show()
```

```python
plt.figure(figsize=(26,5))
plt.title("Adilabad Max Temp")
sns.lineplot(y='Max Temp (°C)',x='Mandal',data=b)
plt.show()
```

```python
plt.figure(figsize=(26,5))
plt.title("Adilabad Min Humidity")
sns.barplot(y='Min Humidity (%)',x='Mandal',data=b)
plt.show()
```

```python
plt.figure(figsize=(26,5))
plt.title("Adilabad Min Humidity")
sns.lineplot(y='Min Humidity (%)',x='Mandal',data=b)
plt.show()
```

```python
plt.figure(figsize=(26,5))
plt.title("Adilabad Max Humidity")
sns.barplot(y='Max Humidity (%)',x='Mandal',data=b)
plt.show()
```

```python
plt.figure(figsize=(26,5))
plt.title("Adilabad Max Humidity")
sns.lineplot(y='Max Humidity (%)',x='Mandal',data=b)
plt.show()
```

```python
plt.figure(figsize=(26,5))
plt.title("Adilabad AVG Temparature in 2021")
sns.barplot(y='AVG',x='Mandal',data=b)
plt.show()
```
