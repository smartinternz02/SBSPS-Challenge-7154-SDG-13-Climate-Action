# MACHINE LEARNING

## Cleaning

```python
data.dtypes
```

```python
frames=[data1,data,data2,data3,data4,data5]
result=pd.concat(frames)
result
```

```python
result.dtypes
result.drop('Date',axis=1,inplace=True)
```

```python
a=result.dtypes==object
obj=a[a==True].index
obj=np.array(obj)
obj
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
for i in obj:
    result[i]=label.fit_transform(result[i])
    result[i]=label.fit_transform(result[i])
```

```python
result
```

```python
result.drop_duplicates(inplace=True)
result
```

```python
result['Rain (mm)'].nunique()
```

```python
X=result.drop('Rain (mm)',axis=1)
X
```

```python
y=result['Rain (mm)']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

## K nearest neighbours

```python
from sklearn.neighbors import KNeighborsRegressor
model1=KNeighborsRegressor(n_neighbors=20)
model1.fit(X_train,y_train)
```

```python
model1.predict(X_test)
print(model1.score(X_train,y_train))
print(model1.score(X_test,y_test))
```

## Decision Trees

```python
from sklearn.tree import DecisionTreeRegressor
model2=DecisionTreeRegressor(max_depth=6)
model2.fit(X_train,y_train)
```

```python
model2.predict(X_test)
print(model2.score(X_train,y_train))
print(model2.score(X_test,y_test))
```

## Converting to classification

```python
result.loc[result['Rain (mm)']>0.0,'Rain (mm)']=1
result
```

```python
result[result['Rain (mm)']==1]
```

```python
result[result['Rain (mm)']==0]
```

```python
X1=result.drop('Rain (mm)',axis=1)
y1=result['Rain (mm)']
```

## down sampling

```python
from imblearn.under_sampling import TomekLinks
under_sampling=TomekLinks()
X1,y1=under_sampling.fit_resample(X1,y1)
```

```python
X1
```

```python
y1
```

```python
y1[y1==0]
```

```python
y1[y1==1]
```

```python
from imblearn.over_sampling import SMOTENC
oversample = SMOTENC(categorical_features=[1,0], random_state = 100)
X1, y1 = oversample.fit_resample(X1, y1)
```

```python
X1
```

```python
from sklearn.model_selection import train_test_split
X_train1,X_test1,y_train1,y_test1=train_test_split(X1,y1,random_state=0)
```

## Random forests

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(max_depth=10)
forest.fit(X_train1,y_train1)
```

```python
forest.predict(X_test1)
print(forest.score(X_train1,y_train1))
print(forest.score(X_test1,y_test1))
```

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
model3=LogisticRegression()
model3.fit(X_train1,y_train1)
```

```python
model3.predict(X_test1)
print(model3.score(X_train1,y_train1))
print(model3.score(X_test1,y_test1))
```
