```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
b=pd.read_excel('Warangal-1.xlsx')
b
```

```python
plt.figure(figsize=(15,5))
plt.title("Warangal Rain")
sns.scatterplot(x='Mandal',y='Rain (mm)',data=b)
plt.show()
```

```python
plt.figure(figsize=(18,5))
plt.title("Warangal Rain")
sns.lineplot(x='Mandal',y='Rain (mm)',data=b)
plt.show()
```
